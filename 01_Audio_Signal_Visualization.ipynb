```python
# 첫 번째 오디오 연구 코드
import librosa
import matplotlib.pyplot as plt

# 샘플 데이터 로드
y, sr = librosa.load(librosa.ex('trumpet'))

# 파형 시각화
plt.figure(figsize=(10, 4))
librosa.display.waveshow(y, sr=sr)
plt.title('My First Audio Analysis - Sj Kim')
plt.show()
```
